```python
x = data.iloc[:,:-1]
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
y = data.iloc[:,-1]

GE = LabelEncoder()
y = GE.fit_transform(y)
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.svm import SVC
from sklearn import metrics
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
svc = SVC(kernel = 'linear')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
svc = SVC(kernel = 'rbf')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
svc = SVC(kernel = 'poly')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
svc = SVC(kernel = 'sigmoid')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print('Accuracy Score:')
print(metrics.accuracy_score(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_classification

# Generate synthetic data 
x,y = make_classification (n_samples = 100, n_features = 2, n_classes=2,n_informative = 2, n_redundant = 0,random_state=42)

# Train the model
clf = svm.SVC(kernel='sigmoid',C=1)
clf.fit(x,y)

# Create a mesh grid
xlim = (x[:,0].min()-1,x[:,0].max() +1)
ylim = (x[:,1].min()-1,x[:,1].max() +1)
xx,yy = np.meshgrid(np.linspace(xlim[0],xlim[1],500), np.linspace(ylim[0],ylim[1],500))

# Compute the decision function
Z = clf.decision_function(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundary
plt.contour(xx,yy,Z,colors=['k'],levels=[0],alpha = 0.5, linestyles= ['--'])
plt.scatter(x[:,0],x[:,1], c=y, edgecolor= 'k', marker= 'x', cmap= plt.cm.Paired)
plt.title('SVM with Linear Kernel')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2 ")
plt.colorbar(label='Class')

plt.show()
```
